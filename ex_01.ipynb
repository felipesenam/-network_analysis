2. Describe how a breadth-first search algorithm works.

3. How many edges does a complete graph with n vertices have? What about a complete directed graph with n vertices?
   
 **R:** Um grafo completo de n vértices, denominado $K_{n}$, é composto por $\frac{n\cdot(n-1)}{2}$ arestas. Um grafo direcionado por sua vez, conseguiria o dobro do valor, já que cada um de seus nós podem ter arestas em ambos os sentidos, isto é: $n\cdot(n-1)$ arestas.

4. What are isomorphic graphs? Draw an example.

 **R:** Grafos isomórficos (isomorphic graphs) são grafos que possuem a mesma estrutura que dispõe seus nós, independentemente dos nomes atribuídos a eles.

```python
G1 = nx.Graph(
    [(i, j) for i in range(5) for j in range(5) if i != j]
)
nx.draw(G1, with_labels=True)
plt.show()

G2 = nx.Graph(
    [(i, j) for i in range(5, 10) for j in range(5, 10) if i != j]
)
nx.draw(G2, with_labels=True)
plt.show()
```

6. Given the digraph G = (V, E) where V = {M, N, O, P, Q, R, S} and
E ={(M, S), (N, O), (P, R), (N, S), (O, M), (N, Q), (O, M), (P, P), (S, M), (O, N),  (S, M), (N, R), (P, M), (M, S)}

    a) Specify, if any, a simple path from vertex M to vertex S.\
    **R:** M e S são vizinhos, e portanto o caminho entre eles é [M, S]

    b) Specify, if any, a simple cycle, involving at least 4 nodes.\
    **R:** Não há ciclos no grafo.

    c) Is the digraph connected or not connected?\
    **R:** O grafo é conectado.

    d) What is the degree of vertices N and R.\
    **R:** 5 e 2, respectivamente.

    e) Represent the digraph using adjacency list representation.\
    **R:**
        M: [S]
        N: [O, S, Q, R]
        O: [M, N]
        P: [M, P, R]
        Q: []
        R: []
        S: [M]

    f) Represent the digraph using adjacency matrix representation.\
    **R:**
            M  N  O  P  Q  R  S
        M   0  0  0  0  0  0  1
        N   0  0  1  0  1  1  1
        O   1  1  0  0  0  0  0
        P   1  0  0  1  0  1  0
        Q   0  0  0  0  0  0  0
        R   0  0  0  0  0  0  0
        S   1  0  0  0  0  0  0

```python
G = nx.DiGraph()
G.add_nodes_from(["M","N","O","P","Q","R","S"])
G.add_edges_from([
    ("M", "S"),
    ("N", "O"),
    ("P", "R"),
    ("N", "S"),
    ("O", "M"),
    ("N", "Q"),
    ("O", "M"),
    ("P", "P"),
    ("S", "M"),
    ("O", "N"),
    ("S", "M"),
    ("N", "R"),
    ("P", "M"),
    ("M", "S")
])

nx.draw(G, with_labels=True)
plt.show()
```

```python
print(G.degree("N"))
print(G.degree("R"))
```

9. Mr. A is friend with Mrs. B, but she doesn't like him back. She has a reciprocal friendship with both C and D, but only C considers D a friend. D has also sent friend requests to E, F, G, and H but, so far, only G replied. G also has a reciprocal relationship with A. Draw the corresponding directed graph.

```python
G = nx.DiGraph([
    ("A", "B"),
    ("B", "C"),
    ("B", "D"),
    ("C", "B"),
    ("D", "B"),
    ("C", "D"),
    ("D", "E"),
    ("D", "G"),
    ("D", "H"),
    ("G", "D"),
    ("A", "G"),
    ("G", "A"),
])

nx.draw(G, with_labels=True)
plt.show()
```

10. Draw the graph from the previous exercise as undirected and weighted, with the weight being 2 if the connection is reciprocal, 1 otherwise.

```python
G = nx.Graph([
    ("A", "B", {"weight":1}),
    ("B", "C", {"weight":2}),
    ("B", "D", {"weight":2}),
    ("C", "D", {"weight":1}),
    ("D", "E", {"weight":1}),
    ("D", "G", {"weight":2}),
    ("D", "H", {"weight":1}),
    ("A", "G", {"weight":2}),
])

nx.draw(G, with_labels=True)
plt.show()
```
