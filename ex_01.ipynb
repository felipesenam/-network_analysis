3. How many edges does a complete graph with n vertices have? What about a complete directed graph with n vertices?

**R:** Um grafo completo de n vértices, denominado $K_{n}$, é composto por $\frac{n\cdot(n-1)}{2}$ arestas. Um grafo direcionado por sua vez, conseguiria o dobro do valor, já que cada um de seus nós podem ter arestas em ambos os sentidos, isto é: $n\cdot(n-1)$ arestas.

4. What are isomorphic graphs? Draw an example.

**R:** Grafos isomórficos (isomorphic graphs) são grafos onde há uma correspondência mútua que estabelece uma aresta entre nós em ambos os sentidos. Isto é: se o nó 5 se liga com o 3 do meu grafo, então meu nó 3 tem que necessariamente se relacionar com o nó 5. A visualização de grafos assim é facilitada quando posta em uma matriz. As arestas são dispostas simetricamente em torno da diagonal principal da matriz.

```python
G = nx.Graph(
    [(i, j) for i in range(5) for j in range(5) if i != j]
)


nx.draw(G)
plt.show()
```

9. Mr. A is friend with Mrs. B, but she doesn't like him back. She has a reciprocal friendship with both C and D, but only C considers D a friend. D has also sent friend requests to E, F, G, and H but, so far, only G replied. G also has a reciprocal relationship with A. Draw the corresponding directed graph.

```python
G = nx.DiGraph([
    ("A", "B"),
    ("B", "C"),
    ("B", "D"),
    ("C", "B"),
    ("D", "B"),
    ("C", "D"),
    ("D", "E"),
    ("D", "E"),
    ("D", "G"),
    ("D", "H"),
    ("G", "D"),
    ("A", "G"),
    ("G", "A"),
])

nx.draw(G)
plt.show()
```
